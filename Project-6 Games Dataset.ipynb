```python
studios.groupby(['Genre','Publisher']).agg(['sum','count','mean','std']).sort_values(('Global_Sales','sum'),ascending=False)
```

```python
studios.groupby(['Genre','Publisher']).agg(['sum','count','mean','std'])\
        .rename({'sum':'total_revenue','count':'no_games'},axis=1)
```

```python
studios.groupby(['Genre','Publisher']).agg(total_revenue=('Global_Sales','sum') ,
                                           no_games=('Global_Sales','count'),
                                           revenue_std=('Global_Sales','std'),
                                           revenue_mean=('Global_Sales','mean')).sort_values('total_revenue',ascending=False)
```
